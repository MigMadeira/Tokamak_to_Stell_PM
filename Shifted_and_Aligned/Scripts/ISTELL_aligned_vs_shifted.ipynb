```python
import matplotlib.pyplot as plt
import os
from pathlib import Path
import numpy as np 
import glob
```

```python
path = os.getcwd()
list_folders = [f.path for f in os.scandir(path) if f.is_dir()]

ISTELL_shifted = np.loadtxt(os.path.join(path,"../Data/Shifted/R2history_K32100_nphi64_ntheta64.txt"),usecols=0)
ISTELL_aligned = np.loadtxt(os.path.join(path,"../Data/Aligned/R2history_K23100_nphi64_ntheta64.txt"),usecols=0)

ISTELL_results_list=[ISTELL_shifted, ISTELL_aligned]

ISTELL_shifted_vol_eff = np.loadtxt(os.path.join(path,"../Data/Shifted/eff_vol_history_K32100_nphi64_ntheta64.txt"),usecols=0)
ISTELL_aligned_vol_eff = np.loadtxt(os.path.join(path,"../Data/Aligned/eff_vol_history_K23100_nphi64_ntheta64.txt"),usecols=0)

ISTELL_vol_eff_list=[ISTELL_shifted_vol_eff, ISTELL_aligned_vol_eff]
```

```python
plt.figure()
plt.xlabel(r'$V_{eff}$',fontsize=16)
plt.ylabel(r'$f_B$ $(T^2m^2)$', fontsize=16)

color_list = ['tab:orange', 'tab:blue']

for i in range(len(ISTELL_results_list)):
        plt.semilogy(ISTELL_vol_eff_list[i], ISTELL_results_list[i], linewidth=2, color = color_list[i])

                
plt.legend(['shifted', 'aligned'], fontsize=14)
plt.grid(which='both')
plt.xlim(0,0.14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.tight_layout()
plt.savefig('../Plots/ISTELL_shifted_vs_aligned_fixed_current.pdf')
```
