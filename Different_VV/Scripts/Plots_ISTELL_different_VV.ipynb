```python
import matplotlib.pyplot as plt
import os
from pathlib import Path
import numpy as np 
import glob
```

```python
path = os.getcwd()
parent_directory = os.path.dirname(path)
path = os.path.join(parent_directory, 'Data')

ISTELL_different_VV = np.loadtxt(os.path.join(path,"eq_cs/R2history_K44000_nphi64_ntheta64.txt"),usecols=0)
ISTELL_different_VV_circular = np.loadtxt(os.path.join(path,"circular_cs/R2history_K27600_nphi64_ntheta64.txt"),usecols=0)
ISTELL_results_list=[ISTELL_different_VV,ISTELL_different_VV_circular]

ISTELL_different_VV_vol_eff = np.loadtxt(os.path.join(path,"eq_cs/eff_vol_history_K44000_nphi64_ntheta64.txt"),usecols=0)
ISTELL_different_VV_circular_vol_eff = np.loadtxt(os.path.join(path,"circular_cs/eff_vol_history_K27600_nphi64_ntheta64.txt"),usecols=0)

ISTELL_vol_eff_list = [ISTELL_different_VV_vol_eff, ISTELL_different_VV_circular_vol_eff]
```

```python
plt.figure(figsize=(6,5))
plt.xlabel(r'$V_{eff}$',fontsize=18)
plt.ylabel(r'$f_B$ $(T^2m^2)$',fontsize=18)


for i in range(len(ISTELL_results_list)):
    plt.semilogy(ISTELL_vol_eff_list[i], ISTELL_results_list[i])
    print(min(ISTELL_results_list[i]))

                
plt.legend(['Equilibrium-shaped VV', 'Circular VV'],fontsize=16)
plt.xticks(np.arange(0,0.181,0.03),fontsize=16)
plt.yticks(fontsize=16)
plt.grid(which='both')
plt.xlim(-0.00005,0.18)
plt.ylim(1e-6,1e-2)
plt.tight_layout()

plt.savefig('../Plots/ISTELL_different_VV_QA.pdf')
```
